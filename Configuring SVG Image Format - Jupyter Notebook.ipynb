```python
from matplotlib import pyplot as plt
import pandas as pd
import seaborn as sns
from sklearn.datasets import load_iris
```

```python
%matplotlib inline

data = load_iris()
cols = data.feature_names
data= data.data

df = pd.DataFrame(data, columns = cols)
```

```python
df.head()
```

#### Use Jupyter Notebook to Configure svg Image Format

```python
# this is the magic command to use
%config InlineBackend.figure_format='svg'

#define subplot
fig, ax = plt.subplots(figsize=(5,5))

# plot now
sns.regplot(x = 'petal width (cm)', y = 'petal length (cm)',
           data = df, fit_reg = False)

plt.show()
```
