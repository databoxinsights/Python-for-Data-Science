```python
# import required libraries
import pandas as pd
import json
```

```python
with open('hotel_reviews.json','r') as json_file:
    json_data = json.load(json_file)
json_file.close()
```

```python
# set up function that returns JSON data to a Pandas dataframe
def cooler_hotel_reviews(data):
    """
    data = expects a json type object like json_data from above
    """
    df = pd.DataFrame(data['Reviews'])
    df_ratings = df.Ratings.dropna().apply(pd.Series) # Ratings is a dictionary witin the json_data, so unpacking here to Columns
    df = (pd
          .concat([df,df_ratings], axis = 1)
          .drop(columns=['Ratings']))
    df.Date = pd.to_datetime(df.Date)
    df.set_index('ReviewID', inplace = True)
    
    return df
    
```
