```python
import pandas as pd
import numpy as np 

# create some fake data with age ranges betwen 1 to 120; make 50 observations 
age = np.linspace(1,120,50)

# just creating an id field; not really needed. 
col_id = np.arange(0,len(age),1)

# creat the bins to be used in pd.cut()
age_bins = np.linspace(0,130,10)

# bring it all together now 
df = pd.DataFrame({'col_id':col_id,'age':age})
df['age_bins'] = pd.cut(df['age'], bins=age_bins)
df.head()
```

```python
new_df = pd.get_dummies(df,columns=['age_bins'])
new_df.drop(columns = ['age'], inplace = True)
new_df.head(25)
```
